```python
import nltk
import pandas as pd
import re
import string
from sklearn.feature_extraction.text import CountVectorizer
# if no stopwords, run: nltk.download('popular')

```

```python
stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()

def count_punct(text):
        count = sum([1 for char in text if char in string.punctuation])
        return round(count/(len(text) - text.count(" ")), 3)*100

#read in data to data frame, replace with your CSV
df = pd.read_csv('IAE1Gd35.csv',encoding = "ISO-8859-1")
df['body_len'] = df['Desc'].apply(lambda x: len(x) - x.count(" "))
df['punct%'] = df['Desc'].apply(lambda x: count_punct(x))

#Remove punctuation, tokenize, remove stopwords, stem words not in stopwords, return list of tokens
def clean_text(text):
        text = "".join([word.lower() for word in text if word not in string.punctuation])
        tokens = re.split('\W+', text)
        text = " ".join([ps.stem(word) for word in tokens if word not in stopwords])  
        return text
    
df['cleaned text'] = df['Desc'].apply(lambda x: clean_text(x))
print (df.head(2))
#count_vect = CountVectorizer(analyzer=clean_text)
#x_counts = count_vect.fit_transform(df['Desc'])
#print (x_counts.shape)
#print (count_vect.get_feature_names())
```
